```python
#check for null values
df.isna().any()
```

```python
#Fill the null values with zeros
df['Not_Outs'] = df['Not_Outs'].fillna(0)
df['Highest_Inns_Scored'] = df['Highest_Inns_Scored'].fillna(0)
df['Hundreds_Scored'] = df['Hundreds_Scored'].fillna(0)
df
```

```python
#split the span column into start date and end date
df['Span'].str.split(pat = '-')
```

```python
df['Start_Date'] = df['Span'].str.split(pat = '-').str[0]
```

```python
df['End_Date'] = df['Span'].str.split(pat = '-').str[1]
```

```python
df = df.drop(['Span'], axis = 1)
df
```

```python
#Split up the country from the player
df['Country'] = df['Player'].str.split(pat = '(').str[1]
df
```

```python
df['Player'] = df['Player'].str.split(pat = '(').str[0]
df
```

```python
df
```

```python
#Remove the ')' from the country names
df['Country'] = df['Country'].str.rstrip(')')
df
```

```python
#df['Balls_Faced'].astype('int') ...remove the + from the data first then convert
df['Balls_Faced'] = df['Balls_Faced'].str.split(pat = '+').str[0]
```

```python
df[df.Balls_Faced == '-']
```

```python
df['Balls_Faced'] = df['Balls_Faced'].replace('-',0)
```

```python
df['Balls_Faced'] = df['Balls_Faced'].astype('int')
```

```python
df['Batting_Strike_Rate'] = df['Batting_Strike_Rate'].astype('int')
```

```python
df['Start_Date'] = df['Start_Date'].astype('int')
df['End_Date'] = df['End_Date'].astype('int')
```

```python
df.dtypes
```

```python
df['Batting_Strike_Rate'] = df['Batting_Strike_Rate'].astype('float')
```

```python
#Build out a career length column
df['Career_Length'] = df['End_Date'] - df['Start_Date']
df
```

```python
#Average career length of cricketers in this data set
df['Career_Length'].mean()
```

```python
#Average strike rate for cricketers who played over 10 years
df[df['Career_Length'] > 10]['Batting_Strike_Rate'].mean()
```
